# Clean Data

**Data cleaning** is an essential step to ensure that the dataset is ready for analysis and free from inconsistencies.<br>
Although this dataset is generally clean and well-structured, some issurs were identified:<br>
1. Columns containing time are stored in integer format(e.g., **50.8**), which is difficult to interpret. These were converted to string format(e.g., **0508** in HHMM format) for better readability.
2. The dataset was then divided into apecific DataFrames for ease of use:
    - **df_flights** : Contains all flight records.
    - **cancelled_flights** : Contains only canceled flight records.
    - **delayed_fligths** : Contains only delayed flight records.

```python
# change time format from int to str
df_flights = pd.read_csv('Dataset/flights.csv', dtype={'DEPARTURE_TIME' : str,
                                                        'WHEELS_OFF' : str, 
                                                        'WHEELS_ON' : str, 
                                                        'SCHEDULED_ARRIVAAL' : str, 
                                                        'ARRIVAL_TIME' : str })

print(df_flights.head())
```

```python
# create dataframe for canceled flights
cancelled_flights = df_flights[df_flights['CANCELLED'] == 1].reset_index()
print(cancelled_flights)
```

```python
# create dataframe for delayed flight
delayed_flights = df_flights[
    df_flights['AIR_SYSTEM_DELAY'].notna() & 
    df_flights['SECURITY_DELAY'].notna() & 
    df_flights['AIRLINE_DELAY'].notna() & 
    df_flights['LATE_AIRCRAFT_DELAY'].notna() & 
    df_flights['WEATHER_DELAY'].notna()
].reset_index(drop=True)

print(delayed_flights)
```

```python
# convert dataframes to csv
df_flights.to_csv("Dataset/df_flights.csv", index=False)
cancelled_flights.to_csv("Dataset/cancelled_flights.csv", index=False)
delayed_flights.to_csv("Dataset/delayed_flights.csv", index=False)
```

# EDA (Exploratory data analysis)

**Visualizations** provide a deeper understanding of the dataset by revealing patterns, trends, and relationships that may not be immediately apparent from raw data.

### Find the distribution of flights by airline
This Pie chart show the dirtribution of all flights by airline in 2015, helping to identify which airline had the highest and lowest flights during 2015.<br><br>
**Key Insights :** Sounthwest Airlines Co. had the highest number of flights in 2015, while Virgin America had the lowest.

```python
# merge 2 columns for IATA code with the actual airline name
df_airline = pd.read_csv('Dataset/airlines.csv')
df_flights_merged = pd.merge(df_flights, df_airline, left_on='AIRLINE', right_on='IATA_CODE', how='left')
df_flights_merged.rename(columns={'AIRLINE_x': 'IATA_CODE', 'AIRLINE_y': 'AIRLINE_NAME'}, inplace=True)

# count the number of flight rows for each airline name
airline_counts = df_flights_merged['AIRLINE_NAME'].value_counts()

# create a pie chart
plt.pie(airline_counts, labels=airline_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Flights by Airline')
plt.show()
```

### Find the total delays by airline and delay type
This stack bar chart shows the total delays by airline and delay type in 2015, highlighting the airlines with the highest overall delays. WN (Southwest Airlines Co.) had the most delays, with a total of 416,124 delays in the year, followed by AA (American Airlines), and DL (Delta Airlines). The chart breaks down the delays by type, providing insights into the specific causes behind these delays, such as late aircraft delay, airline delay, and air system delay.<br><br>
Additionally, the size of each airline's fleet can be a contributing factor to the number of delays, as airlines with larger fleets have more flights and may experience a higher total number of delays.<br><br>
**Key Insights :** Southwest Airlines Co. (WN) had the highest number of total delays in 2015, significantly outperforming other airlines. American Airlines (AA) and Delta Airlines (DL) followed.

```python
# List of columns related to delays
delay_columns = ['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']

# Convert the delay data to 0 (no delay) or 1 (delay exists)
delay_counts = delayed_flights[delay_columns].gt(0).astype(int)

# Include the AIRLINE column in the delay_counts DataFrame
delay_counts['AIRLINE'] = delayed_flights['AIRLINE']

# Group by airline and sum the delays
grouped_data = delay_counts.groupby('AIRLINE').sum()

# Calculate the total number of delays for each airline
grouped_data['TOTAL_DELAYS'] = grouped_data.sum(axis=1)

# Sort the data by total delays in descending order
grouped_data = grouped_data.sort_values('TOTAL_DELAYS', ascending=False)

# Reset the index to make 'AIRLINE' a column again
grouped_data = grouped_data.reset_index()

# Create a stacked bar chart
airlines = grouped_data['AIRLINE']
delay_types = grouped_data.drop(columns=['AIRLINE', 'TOTAL_DELAYS'])
ax = delay_types.plot(kind='bar', stacked=True, figsize=(10, 7))

# Set the title and labels
plt.title('Total Delays by Airline and Delay Type')
plt.xlabel('Airline')
plt.ylabel('Number of Delays')
plt.xticks(range(len(airlines)), airlines, rotation=90)
plt.legend(title='Delay Type')

# Add total delays on top of each bar
for i, total_delay in enumerate(grouped_data['TOTAL_DELAYS']):
    ax.text(i, total_delay + 0.5, str(total_delay), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

### Find the delay rate by airline
This bar chart shows the delay rate by airline in 2015, sorted from lowest to highest. The delay rate is calculated as the percentage of delayed flights out of the total flights operated by each airline.<br><br>
**Key Insights :**  Hawaiian Airlines (HA) had the lowest delay rate in 2015, indicating strong operational efficiency with only 11% of flights delayed. In contrast, Spirit Airlines (NK) had the highest delay rate at 29%.

```python
# Calculate the total number of flights per airline
total_flights_per_airline = df_flights.groupby('AIRLINE').size()

# Calculate the number of delayed flights per airline
delayed_flights_per_airline = delayed_flights.groupby('AIRLINE').size()

# Merge the two DataFrames (total flights and cancelled flights)
delay_flights_rate = pd.DataFrame({
    'total_flights': total_flights_per_airline,
    'delayed_flights': delayed_flights_per_airline
})

# Calculate the cancellation rate per airline
delay_flights_rate['delay_rate'] = (delay_flights_rate['delayed_flights'] / delay_flights_rate['total_flights']) * 100

print(delay_flights_rate.sort_values('delay_rate'))
```

```python
# Sort the DataFrame by delay_rate in ascending order
delay_flights_rate = delay_flights_rate.sort_values('delay_rate', ascending=True)

plt.figure(figsize=(10, 6))
plt.bar(delay_flights_rate.index, delay_flights_rate['delay_rate'], width=0.8)

# Add label and title
plt.title('Delay Rate by Airline')
plt.xlabel('Airline')
plt.ylabel('Delay Rate (%)')

plt.show()
```

### Find the cancellation rate by airline
This bar chart shows the cancellation rate by airline in 2015, sorted from lowest to highest. The cancellation rate is calculated as the percentage of canceled flights out of the total flights operated by each airline.<br><br>
**Key Insights :** Hawaiian Airlines (HA) had the lowest cancellation rate in 2015, with just 0.2% of flights canceled, indicating a high level of reliability. On the other hand, Envoy Air (MQ) had the highest cancellation rate at 5%.

### Find the relation between sistance and arrival delay
This scatter plot shows the relationship between the flight distance and arrival delay (in minutes), using a 20% sample of the dataset to reduce density and improve clarity.<br><br>
**Key Insights :** The scatter plot reveals a trend where longer flights generally experience lower arrival delays. This suggests that longer routes might have more efficient scheduling, fewer delays due to air traffic, or better management of time buffers compared to shorter flights.

```python
x = delayed_flights['DISTANCE']  # Distance
y = delayed_flights['ARRIVAL_DELAY']  # Arrival Delay
data = pd.DataFrame({'Distance': x, 'Arrival Delay': y})

# Sample 20% of the data to reduce the density
data_sampled = data.sample(frac=0.2, random_state=42)

# Create plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Distance'], data['Arrival Delay'], alpha=0.2, color='blue')

# Add label and title
plt.title('Relationship Between Distance and Arrival Delay')
plt.xlabel('Distance (miles)')
plt.ylabel('Arrival Delay (minutes)')

# Show plot
plt.tight_layout()
plt.show()
```

### Find the delayed and canceled flights per month
This line chart shows the number of delayed and canceled flights for each month from January to December , alongside the total number of flights in each month.<br><br>
**Key Insights :** The months of May to July and November to December had the highest number of flights, and these months also saw a significant number of delayed flights, which could be attributed to increased travel demand and seasonal factors. February had a particularly high cancellation rate, despite having the lowest total number of flights. This suggests that while February had fewer flights overall, factors such as weather disruptions may have led to a higher proportion of cancellations.

```python
# delayed flights
x_delayed = delayed_flights['MONTH']
y_delayed = delayed_flights['MONTH'].value_counts().sort_index()

# canceled flights
x_cancelled = cancelled_flights['MONTH']
y_cancelled = cancelled_flights['MONTH'].value_counts().sort_index()

# total flight in 2015
y_total = df_flights['MONTH'].value_counts().sort_index()

fig, ax = plt.subplots()

# Plot the delayed, cancelled, and total flights
ax.plot(y_total.index, y_total.values, label='Total Flights', color='green', marker='s', linestyle='--')
ax.plot(y_delayed.index, y_delayed.values, label='Delayed Flights', color='blue', marker='o')
ax.plot(y_cancelled.index, y_cancelled.values, label='Cancelled Flights', color='red', marker='x')

# Add label and title
ax.set(xlabel='Month', ylabel='Number of Flights', title='Delayed, Canceled and Total Flights per Month')

ax.grid()

# Add a legend to differentiate the lines
ax.legend()

plt.show()
```
