### Column details
**YEAR** = Year of the Flight Trip <br>
**MONTH** = Month of the Flight Trip <br>
**DAY** = Day of the Flight Trip <br>
**DAY_OF_WEEK** = Day of week of the Flight Trip <br>
**AIRLINE** = Airline Identifier <br>
**FLIGHT_NUMBER** = Flight Identifier <br>
**TAIL_NUMBER** = Aircraft Identifier <br>
**ORIGIN_AIRPORT** = Starting Airport <br>
**DESTINATION_AIRPORT** = Destination Airport <br>
**SCHEDULED_DEPARTURE** = Planned Departure Time <br>
**DEPARTURE_TIME** = WHEEL_OFF - TAXI_OUT <br>
**DEPARTURE_DELAY** = Total Delay on Departure <br>
**TAXI_OUT** = The time duration elapsed between departure from the origin airport gate and wheels off <br>
**WHEELS_OFF** = The time point that the aircraft's wheels leave the ground <br>
**SCHEDULED_TIME** = Planned time amount needed for the flight trip <br>
**ELAPSED_TIME** = AIR_TIME+TAXI_IN+TAXI_OUT <br>
**AIR_TIME** = The time duration between wheels_off and wheels_on time <br>
**DISTANCE** = Distance between two airports <br>
**WHEELS_ON** = The time point that the aircraft's wheels touch on the ground <br>
**TAXI_IN** = The time duration elapsed between wheels-on and gate arrival at the destination airport <br>
**SCHEDULED_ARRIVAL** = Planned arrival time <br>
**ARRIVAL_TIME** = WHEELS_ON+TAXI_IN <br>
**ARRIVAL_DELAY** = ARRIVAL_TIME-SCHEDULED_ARRIVAL <br>
**DIVERTED** = Aircraft landed on airport that out of schedule <br>
**CANCELLED** = Flight Cancelled (1 = cancelled) <br>
**CANCELLATION_REASON** = Reason for Cancellation of flight: A - Airline/Carrier, B - Weather, C - National Air System, D - Security <br>
**AIR_SYSTEM_DELAY** = Delay caused by air system <br>
**SECURITY_DELAY** = Delay caused by security <br>
**AIRLINE_DELAY** = Delay caused by the airline <br>
**LATE_AIRCRAFT_DELAY** = Delay caused by aircraft <br>
**WEATHER_DELAY** = Delay caused by weather <br>

# Clean Data

```python
# change time format from int to str
df_flights = pd.read_csv('Dataset/flights.csv', dtype={'DEPARTURE_TIME' : str,
                                                        'WHEELS_OFF' : str, 
                                                        'WHEELS_ON' : str, 
                                                        'SCHEDULED_ARRIVAAL' : str, 
                                                        'ARRIVAL_TIME' : str })

print(df_flights.head())
```

```python
# create dataframe for canceled flights
cancelled_flights = df_flights[df_flights['CANCELLED'] == 1].reset_index()
print(cancelled_flights)
```

```python
# create dataframe for diverted flights
diverted_flights = df_flights[df_flights['DIVERTED'] == 1].reset_index()
print(diverted_flights)
```

```python
# create dataframe for delayed flight
delayed_flights = df_flights[
    df_flights['AIR_SYSTEM_DELAY'].notna() & 
    df_flights['SECURITY_DELAY'].notna() & 
    df_flights['AIRLINE_DELAY'].notna() & 
    df_flights['LATE_AIRCRAFT_DELAY'].notna() & 
    df_flights['WEATHER_DELAY'].notna()
].reset_index(drop=True)

print(delayed_flights)
```

# EDA (Exploratory data analysis)

### Find the distribution of flights by airline

```python
# merge 2 columns for IATA code with the actual airline name
df_airline = pd.read_csv('Dataset/airlines.csv')
df_flights_merged = pd.merge(df_flights, df_airline, left_on='AIRLINE', right_on='IATA_CODE', how='left')
df_flights_merged.rename(columns={'AIRLINE_x': 'IATA_CODE', 'AIRLINE_y': 'AIRLINE_NAME'}, inplace=True)

# count the number of flight rows for each airline name
airline_counts = df_flights_merged['AIRLINE_NAME'].value_counts()

# create a pie chart
plt.pie(airline_counts, labels=airline_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Flights by Airline')
plt.show()
```

### Flight Dalays and Cancellation

Overall Delay Distribution

```python
# List of columns related to delays
delay_columns = ['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']

# Convert the delay data to 0 (no delay) or 1 (delay exists)
delay_counts = delayed_flights[delay_columns].gt(0).astype(int)

# Include the AIRLINE column in the delay_counts DataFrame
delay_counts['AIRLINE'] = delayed_flights['AIRLINE']

# Group by airline and sum the delays
grouped_data = delay_counts.groupby('AIRLINE').sum()

# Calculate the total number of delays for each airline
grouped_data['TOTAL_DELAYS'] = grouped_data.sum(axis=1)

# Sort the data by total delays in descending order
grouped_data = grouped_data.sort_values('TOTAL_DELAYS', ascending=False)

# Reset the index to make 'AIRLINE' a column again
grouped_data = grouped_data.reset_index()

# Create a stacked bar chart
airlines = grouped_data['AIRLINE']
delay_types = grouped_data.drop(columns=['AIRLINE', 'TOTAL_DELAYS'])
ax = delay_types.plot(kind='bar', stacked=True, figsize=(10, 7))

# Set the title and labels
plt.title('Delays by Airline and Delay Type')
plt.xlabel('Airline')
plt.ylabel('Number of Delays')
plt.xticks(range(len(airlines)), airlines, rotation=90)
plt.legend(title='Delay Type')

# Add total delays on top of each bar
for i, total_delay in enumerate(grouped_data['TOTAL_DELAYS']):
    ax.text(i, total_delay + 0.5, str(total_delay), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

Scatter plot to show relation between Distance and Arrival delay

```python
x = delayed_flights['DISTANCE']  # Distance
y = delayed_flights['ARRIVAL_DELAY']  # Arrival Delay
data = pd.DataFrame({'Distance': x, 'Arrival Delay': y})

# Sample 20% of the data to reduce the density
data_sampled = data.sample(frac=0.2, random_state=42)

# Create plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Distance'], data['Arrival Delay'], alpha=0.2, color='blue')

# Add label and title
plt.title('Relationship Between Distance and Arrival Delay')
plt.xlabel('Distance (miles)')
plt.ylabel('Arrival Delay (minutes)')

# Show plot
plt.tight_layout()
plt.show()
```
